```python
import pandas as pd

# Load datasets
df1 = pd.read_csv("../datasets/raw/1429_1.csv") # 21 columns
# df1.info()
# df1.head()
df2 = pd.read_csv("../datasets/raw/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv")# 24 columns
# df2.info()
# df2.head()
df3 = pd.read_csv("../datasets/raw/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv")# 24 columns
# df3.info()
# df3.head()

```

After reviewing all three datasets, it is evident that the column structures are largely consistent across the files. However, there are minor variations in the column sequence, along with a few additional columns in some datasets and missing columns in others. Therefore, the data from the three files must be merged carefully to ensure proper compatibility and alignment of corresponding fields.

```python
# 2) Align columns (union of all columns)
all_cols = sorted(set(df1.columns) | set(df2.columns) | set(df3.columns))
df1, df2, df3 = [d.reindex(columns=all_cols) for d in (df1, df2, df3)]

# 3) Combine rows (more training examples)
df = pd.concat([df1, df2, df3], ignore_index=True)

# df.info()
# df.head()
```

```python
# text normalization: remove extra whitespace, newlines, tabs, etc.
import re

def clean_text(text):
    text = str(text)
    text = re.sub(r"\s+", " ", text)  # remove extra whitespace
    text = text.strip()
    return text

df["reviews.text"] = df["reviews.text"].apply(clean_text)
```

```python
# check for conflicting reviews: same text but different ratings
conflicts = df.groupby("reviews.text")["reviews.rating"].nunique()
conflicting_texts = conflicts[conflicts > 1]

print("Conflicting texts:", len(conflicting_texts))
```

```python
# find conflicting texts
conflicts = df.groupby("reviews.text")["reviews.rating"].nunique()
conflicting_texts = conflicts[conflicts > 1].index

# filter full rows that contain those conflicting texts
conflicting_df = df[df["reviews.text"].isin(conflicting_texts)].copy()

print("Total conflicting review texts:", len(conflicting_texts))
print("Total conflicting rows:", len(conflicting_df))


```

```python
#Display conflicts one-by-one (manual inspection)
for text in conflicting_texts:
    subset = df[df["reviews.text"] == text][["reviews.text", "reviews.rating", "reviews.title", "categories", "name"]]
    print("="*120)
    print("Review text:\n", text)
    print("\nRatings found:", subset["reviews.rating"].unique())
    print(subset)

```
